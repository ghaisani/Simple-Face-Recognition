### Simple Face Recognition

Data set from Face Recognition Data, University of Essex, UK
(http://cswww.essex.ac.uk/mv/allfaces/index.html)

```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
cv_img = []
i = 0
j = 1
for img in glob.glob("C:/Users/ASUS/Documents/SISREK/dataset/Train/*/*.jpg"):
    n = cv2.imread(img)
    
    # preprocessing jadi grayscale
    n = cv2.cvtColor( n, cv2.COLOR_RGB2GRAY )
    cv_img.append(n)
    
    # show image
    plt.axis("off")
    plt.imshow(n)
    plt.show()
```

```python
# ekstrasi fitur
F = np.zeros((25,3))
i = 0
j = 1
while i<25:
    F[i] = [np.mean(cv_img[i]),np.std(cv_img[i]),j]
    if (i+1) % 5 == 0:
        j += 1
    i += 1
F
```

```python
# testing

t = cv2.imread("C:/Users/ASUS/Documents/SISREK/dataset/Testing/kaknig/kaknig.10.jpg")

print('Image sebagai testing : ')
plt.axis("off")
plt.imshow(t)
plt.show()

t = cv2.cvtColor( t, cv2.COLOR_RGB2GRAY )
Te = [np.mean(t),np.std(t)]
```

```python
# klasifikasi

dist = np.zeros((len(F),1))
for i in range(len(F)):
    dist[i] = np.sum(abs( (F[i,0]-Te[0]) + (F[i,1]-Te[1]) ))
dist
```

```python
# Hasil klasifikasi
x = np.argmin(dist)
print(x)
print('Jadi gambar kaknig.10 termasuk ke dalam kelas : ', F[x,2])
```
